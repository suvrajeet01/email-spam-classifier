```python
#importing native python modules
import os
from collections import Counter
```

```python
folder = 'email/'
```

```python
#passing folder as a parameter to the listdir function of os module returns all the files inside the folder in form of a list 
files = os.listdir(folder)
files
```

```python
len(files)
```

```python
#displaying output as complete path for each files or emails
emails = [folder + file for file in files]
#emails stores the complete path of each file
emails
```

```python
len(emails)
```

```python
words=[]

for email in emails:
    f=open(email,encoding = 'latin-1')#opening  file
    e = f.read()
    #reads the file that is passed as parameter and stores all the conten in form of a list
    #e stores all the contents of first file in string format
    words+=e.split(' ')
    #converts the above string to list and list is concatenated to words
    #words = words+e.split()
    #words contains a list of all the words from the first file and the loop goes on until the process is completed for all the files in the folder
```

```python
words
```

```python
len(words)
```

```python
#removing all the words that are obstacles for this analysis
#spaces,alphanumeric words

#loop to make a alphanumeric word to a blank space so that we are left only with pure alphabetic words
for i in range(len(words)):
    if not words[i].isalpha():
        words[i]=''
```

```python
#counter is a class to which if a list is passed with huge number of items it returns with a frequency count of each items
word_dic=Counter(words)
#frequency of each items in worlds list in form of a dictionary
word_dic
```

```python
#number of key value pairs
len(word_dic)
```

```python
#segregating only 3000 most frequent words out of 40577 for this particular analysis

#removing the most common and useless key value pair i.e. the blank spaces
del word_dic['']
word_dic = word_dic.most_common(3000)
word_dic
#a tuple
```

```python
#so now according to analysis upto this point it is very clear that the emails conatining these words can be flagged as spam
len(word_dic)
```

```python
for i in word_dic:
    print(i)
#tuple with words and freqeuncy both
```

```python
for i in word_dic:
    print(i[0])
#tuple with only words
```

```python
#obtaining tabular data from the abobe analysis report till now containing rows as each emails with the frequency of word count corresponding to each column and columns as each and every elements of word_dic
#so there will be 3000 columns and 5172 rows in the table

features = []
labels = []

for email in emails:
    f=open(email,encoding='latin-1')
    e=f.read().split(' ')
    data = []
    for i in word_dic:
        data.append(e.count(i[0]))
    features.append(data)
        
    if 'spam' in email:
        labels.append(1)
    if 'ham' in email:
        labels.append(0)
```

```python
len(features)
```

```python
len(labels)
```

```python
import numpy as np
```

```python
features = np.array(features)
features.shape
```

```python
labels=np.array(labels)
labels.shape
```

```python
#now finally the analysed dataset is complete and now the data is to be splitted to training set and test dataset
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,labels,test_size=0.2, random_state = 9)
X_train.shape
```

```python
#from 5172 approximately 1000 data is for test purpose and the remaining 4000 spots for training purpose
#using naive bayes algorithm[based on bayes theorem on probability(conditional probability)] as for this condition this algorithm is best with a accuracy score of about 94%
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
#classifier is an object for Multinomial naive bayes
```

```python
#training the train data
classifier.fit(X_train,y_train)
```

```python
#calcutaing accuracy for the model
y_pred=classifier.predict(X_test)
```
